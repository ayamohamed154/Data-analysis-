## **1) What are the monthly sales trends over the years? Are there specific months with peak sales?**

```python
# Convert 'Order Date' to datetime format with dayfirst=True
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
# Extract year and month
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
# Group by year and month, summing sales
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
# Create a 'Date' column for plotting
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Date'], monthly_sales['Sales'], marker='o')
plt.title('Monthly Sales Trends Over the Years')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
df.to_csv('Cleaned_Superstore.csv', index=False)
```

## **2) Which product categories generate the highest revenue? How do different sub-categories compare within those categories?**

```python
# Group by category and sub-category, summing sales
category_sales = df.groupby(['Category', 'Sub-Category'])['Sales'].sum().reset_index()
# Get total sales per category for comparison
total_category_sales = category_sales.groupby('Category')['Sales'].sum().reset_index()
# Sort categories by total sales
total_category_sales = total_category_sales.sort_values(by='Sales', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(total_category_sales['Category'], total_category_sales['Sales'], color='skyblue')
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group by category and sub-category, summing sales
category_sales = df.groupby(['Category', 'Sub-Category'])['Sales'].sum().reset_index()
# Sort sub-categories by sales within each category
category_sales = category_sales.sort_values(by=['Category', 'Sales'], ascending=[True, False])
# Print the sales by sub-category for each category as a table
print("Sales by Sub-Category within each Category:")
display(category_sales)  # Use display for Jupyter Notebooks
# Create a lined scatter plot
plt.figure(figsize=(12, 6))
for category in category_sales['Category'].unique():
    subset = category_sales[category_sales['Category'] == category]
    plt.plot(subset['Sub-Category'], subset['Sales'], marker='o', label=category)
plt.title('Sales by Sub-Category within Each Category (Lined Scatter Plot)')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Group by segment and sum the sales
segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()
# Display the sales by customer segment as a table
print("Sales by Customer Segment:")
display(segment_sales)  # Use this for Jupyter Notebooks
most_profitable_segment = segment_sales.iloc[0]
print(f"\nThe most profitable segment is '{most_profitable_segment['Segment']}' with total sales of {most_profitable_segment['Sales']:.2f}")
```

## **3) What is the impact of shipping modes (e.g., Second Class, First Class) on delivery times and sales volume?**

```python
# Convert 'Order Date' and 'Ship Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
# Calculate delivery time in days
df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
# Group by shipping mode and calculate total sales and average delivery time
shipping_analysis = df.groupby('Ship Mode').agg(Total_Sales=('Sales', 'sum'),Average_Delivery_Time=('Delivery Time', 'mean')).reset_index()
# Sort by total sales
shipping_analysis = shipping_analysis.sort_values(by='Total_Sales', ascending=False)
# Print the analysis as a table
print("Shipping Mode Analysis:")
display(shipping_analysis)
# Plotting Total Sales and Average Delivery Time
fig, ax1 = plt.subplots(figsize=(10, 6))
# Create a bar plot for total sales
ax1.bar(shipping_analysis['Ship Mode'], shipping_analysis['Total_Sales'], color='lightblue', label='Total Sales')
ax1.set_xlabel('Shipping Mode')
ax1.set_ylabel('Total Sales', color='lightblue')
ax1.tick_params(axis='y', labelcolor='lightblue')
# Create a second y-axis for average delivery time
ax2 = ax1.twinx()
ax2.plot(shipping_analysis['Ship Mode'], shipping_analysis['Average_Delivery_Time'], color='orange', marker='o', label='Average Delivery Time')
ax2.set_ylabel('Average Delivery Time (Days)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
# Add titles and legends
plt.title('Impact of Shipping Modes on Delivery Times and Sales Volume')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

## **4) Which states or regions contribute the most to total sales? Are there regions with notably low sales?**

```python
# Group by state and region, summing sales
state_sales = df.groupby(['State', 'Region'])['Sales'].sum().reset_index()
# Sort by total sales
state_sales = state_sales.sort_values(by='Sales', ascending=False)
# Print the sales by state and region as a table
print("Sales by State and Region:")
display(state_sales)
# Identify regions with notably low sales
low_sales_threshold = state_sales['Sales'].quantile(0.25)  # You can adjust this threshold
low_sales = state_sales[state_sales['Sales'] < low_sales_threshold]
# Display states with notably low sales
print("States with Notably Low Sales:")
display(low_sales)
# Filter out states with notably low sales from the plotting data
filtered_state_sales = state_sales[~state_sales['State'].isin(low_sales['State'])]
# Plotting total sales by state excluding low sales
plt.figure(figsize=(14, 7))
plt.bar(filtered_state_sales['State'], filtered_state_sales['Sales'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by State (Excluding Low Sales)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## **5) What is the average time taken for each category from order date to ship date?**

```python
# Convert 'Order Date' and 'Ship Date' to datetime format, handling errors
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True, errors='coerce')
# Drop rows with NaT in either date column
df = df.dropna(subset=['Order Date', 'Ship Date'])
# Calculate delivery time in days
df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
# Calculate the average delivery time
average_delivery_time = df['Delivery Time'].mean()
print(f"Average Delivery Time: {average_delivery_time:.2f} days")
# Group by category and calculate average delivery time
category_delivery_time = df.groupby('Category')['Delivery Time'].mean().reset_index()
# Sort by average delivery time
category_delivery_time = category_delivery_time.sort_values(by='Delivery Time', ascending=False)
# Print average delivery time by category
print("Average Delivery Time by Category:")
display(category_delivery_time)
plt.figure(figsize=(10, 8))
plt.pie(category_delivery_time['Delivery Time'], labels=category_delivery_time['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Average Delivery Time by Category')
plt.axis('equal')
plt.show()
```

## **8) Does we have customers repeated their visits?**

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=False)  # Adjust 'dayfirst' based on your date format
# Drop duplicates of the same 'Order Date' for the same 'Customer ID'
unique_transactions = df[['Customer ID', 'Customer Name', 'Order Date']].drop_duplicates()
# Count the number of unique transactions for each customer
transaction_count = unique_transactions.groupby(['Customer ID', 'Customer Name']).size().reset_index(name='Transaction Count')
# Sort by number of transactions (optional: limit the plot to top customers for readability)
transaction_count = transaction_count.sort_values(by='Transaction Count', ascending=False).head(10)
# Plotting the results
plt.figure(figsize=(10, 6))
sns.barplot(x='Transaction Count', y='Customer Name', data=transaction_count, palette='viridis')
plt.title('Top 10 Customers by Number of Unique Transactions')
plt.xlabel('Number of Transactions')
plt.ylabel('Customer Name')
plt.show()
```

## **9) Lowest Repeated Purchased Products**

```python
# Identify repeat customers
repeat_customers = df[df.duplicated(subset='Customer ID', keep=False)]
# Aggregate sales by product for repeat customers
repeat_product_sales = repeat_customers.groupby('Product Name')['Sales'].sum().reset_index()
# Count total purchases per product
total_product_sales = df.groupby('Product Name')['Sales'].sum().reset_index()
# Merge the two DataFrames to calculate repeat purchase rate
product_sales = pd.merge(total_product_sales, repeat_product_sales, on='Product Name', how='left', suffixes=('_Total', '_Repeat'))
# Fill NaN values with 0 for products with no repeat sales
product_sales['Sales_Repeat'] = product_sales['Sales_Repeat'].fillna(0)
# Calculate repeat purchase rate
product_sales['Repeat_Purchase_Rate'] = (product_sales['Sales_Repeat'] / product_sales['Sales_Total']) * 100
# Filter out products with a 100% repeat purchase rate
product_sales = product_sales[product_sales['Repeat_Purchase_Rate'] < 100]
# Sort by repeat purchase rate to find the lowest
lowest_loyalty_products = product_sales.sort_values(by='Repeat_Purchase_Rate').head(10)
# Display the products with the lowest repeat purchase rates
print("Products with the Lowest Repeat Purchase Rates:")
display(lowest_loyalty_products[['Product Name', 'Repeat_Purchase_Rate']])
plt.figure(figsize=(10, 6))
plt.barh(lowest_loyalty_products['Product Name'], lowest_loyalty_products['Repeat_Purchase_Rate'], color='lightcoral')
plt.xlabel('Repeat Purchase Rate (%)')
plt.title('Top Products with Lowest Repeat Purchase Rates')
plt.xlim(0, 100)
plt.show()
```
